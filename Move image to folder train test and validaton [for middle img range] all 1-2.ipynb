```python
# Train
import pandas as pd
dataframe = pd.read_csv('/home/yupaporn/code/EffNet-Person-Classify/data_label_by_tan_09102023_splited_imgpath28_train_middleframe.csv') #‡πÅ‡∏Å‡πâ data ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏ï‡∏≤‡∏° fold
# dataframe = dataframe[dataframe['img_no'] ==1]
dataframe = dataframe.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
print(dataframe.shape)
dataframe.head()
```

```python
# Train
import pandas as pd
dataframe1 = pd.read_csv('/home/yupaporn/code/EffNet-Person-Classify/data_label_by_tan_08102023_splited_imgpath28_train.csv') #‡πÅ‡∏Å‡πâ data ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏ï‡∏≤‡∏° fold
# dataframe = dataframe[dataframe['img_no'] ==1]
dataframe1 = dataframe1.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
print(dataframe1.shape)
dataframe1.head()
```

```python
dataframe = dataframe.append(dataframe1)
print(dataframe.shape)
```

```python
# path_men = dataframe['path_img_28'].tolist()
# path_men
```

```python
img_no_df = dataframe[dataframe['Path_folder_web'] =='']
paths= list(set(dataframe['Path_folder_web']))
for i in range(len(paths)):
    dff = dataframe[dataframe['Path_folder_web'] ==paths[i]]
    image_no = int(len(dff)/2)
    img_no = dff[dff['img_no'] == image_no]
    img_no_df = img_no_df.append(img_no)
    
print(img_no_df.shape)
img_no_df.head()

path_men = img_no_df['path_img_28'].tolist()
```

```python
import shutil
import os
path_men_new = []
img_new = []
train_dir = '/media/tohn/HDD/BearHouse-Person-Data/data_R3_middleframe/train/'
for i in range(len(path_men)):
    filename = str(i) + '_' + path_men[i].split('/')[-1]
    img_new.append(filename)
    filename_path = train_dir + filename
    path_men_new.append(filename_path)
    # moving image
    dst = os.path.join(train_dir,filename)
    shutil.copyfile(path_men[i], dst)
```

```python
img_no_df['path_img_28_train'] = path_men_new
img_no_df['img_28_train'] = img_new
img_no_df.to_csv('/home/yupaporn/code/EffNet-Person-Classify/data_label_by_tan_09102023_splited_imgpath28_train_middleframe.csv')
```
